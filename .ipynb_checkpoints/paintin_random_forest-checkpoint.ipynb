```python
from scipy import ndimage, misc
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
```

```python
from skimage.color import rgb2gray

img = misc.imread('image.jpg')
plt.imshow(img)

x = len(img[0])
y = len(img)
pixel_num = x * y
```

```python
img_gray = rgb2gray(img)

plt.imshow(rgb2gray(img), cmap = 'gray')
```

```python
def sample_pixels(x, y):

    random_pixels = []

    while len(random_pixels) < 5000:
        x_random = np.random.randint(0, x)
        y_random = np.random.randint(0, y)

        temp = [x_random, y_random]

        if temp not in random_pixels:
            random_pixels.append(temp)
    
    return random_pixels
```

```python
from sklearn.preprocessing import normalize

def preprocess_image(img):
    img = normalize(img)
    return img
```

```python
#prp_image = preprocess_image(img)
# Gives dimension error
```

```python
X = sample_pixels(x, y)
```

```python
def return_value(pixel, img):
    
    x_coordinate = pixel[0]
    y_coordinate = pixel[1]
    
    return img[y_coordinate][x_coordinate]
```

```python
def create_y(samples, img):
    y = []
    for pixel in samples:
        rgb = return_value(pixel, img)
        y.append(rgb)
    
    return y
```

```python
Y = create_y(X, img_gray)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression

regr = RandomForestRegressor(max_depth=5, random_state=0, n_estimators=100)
regr.fit(X, Y)
```

```python
regr.predict([[0,1]])
```

```python
def approximate_image(regr):
    app_image = []
    
    for i in range(100):
        line = []
        for j in range(100):
            line.append(regr.predict([[j,i]]))
        app_image.append(line)
        
    return app_image
```

```python
app_image = approximate_image(regr)
```

```python
app_image = np.asarray(app_image)
```

```python
plt.imshow(X = app_image)
```
